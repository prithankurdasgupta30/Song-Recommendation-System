```python
import numpy as mp
import pandas as pd
```

```python
songs = pd.read_csv("Spotify Most Streamed Songs.csv")
```

```python
songs.head()
```

```python
songs = songs[['track_name', 'artist(s)_name','released_year','danceability_%','energy_%','acousticness_%','instrumentalness_%','liveness_%','speechiness_%']]
songs.head()
```

```python
songs.isnull().sum()
```

```python
def convert(text):
    L = []
    for i in text.split(','):
        L.append(i.replace(" ",""))
    return L
```

```python
songs['artist(s)_name'] = songs['artist(s)_name'].apply(convert)
songs.head()
```

```python
songs['released_year'] = songs['released_year'].apply(lambda x: [x] if pd.notnull(x) else [])
```

```python
songs['danceability_%'] = songs['danceability_%'].apply(lambda x: [x] if pd.notnull(x) else [])
```

```python
songs['energy_%'] = songs['energy_%'].apply(lambda x: [x] if pd.notnull(x) else [])
```

```python
songs['acousticness_%'] = songs['acousticness_%'].apply(lambda x: [x] if pd.notnull(x) else [])
```

```python
songs['instrumentalness_%'] = songs['instrumentalness_%'].apply(lambda x: [x] if pd.notnull(x) else [])
```

```python
songs['liveness_%'] = songs['liveness_%'].apply(lambda x: [x] if pd.notnull(x) else [])
```

```python
songs['speechiness_%'] = songs['speechiness_%'].apply(lambda x: [x] if pd.notnull(x) else [])
```

```python
songs.head()
```

```python
songs['tags'] = songs['artist(s)_name'] + songs['released_year'] + songs['danceability_%'] + songs['energy_%'] + songs['acousticness_%'] + songs['instrumentalness_%'] + songs['liveness_%'] + songs['speechiness_%']
```

```python
songs.head()
```

```python
df = songs[['track_name','tags']]
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda tag_list: [i.lower() if isinstance(i, str) else i for i in tag_list])
```

```python
df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(tag_list):
    y = []
    for i in tag_list:
        if isinstance(i, str):
            y.append(ps.stem(i.lower()))
        else:
            y.append(i)
    return " ".join(map(str, y))
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
def recommend(songs):
  songs_index = df[df['track_name'] == songs].index[0]
  distances = similarity[songs_index]
  songs_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:21]
  for i in songs_list:
    print(df.iloc[i[0]].track_name)
```

```python
recommend('Late Night Talking')
```
